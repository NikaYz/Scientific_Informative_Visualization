2.  LGA Code Based Attrion Rate across different years
---


```python
lga = pd.read_csv("LGA HE Attrition Data, 2005-2013.csv")
lga.head(10)
```

```python
df_melted = lga.melt(id_vars=["LGA_code", "LGA_name"], var_name="Year", value_name="Attrition_Rate")

df_melted['Attrition_Rate'] = df_melted['Attrition_Rate'].replace({'-': None, '': None})
df_melted['Attrition_Rate'] = pd.to_numeric(df_melted['Attrition_Rate'], errors='coerce')

df_melted.dropna(subset=['Attrition_Rate'], inplace=True)

grouped_df = df_melted.groupby(['LGA_name', 'Year'], as_index=False)['Attrition_Rate'].sum()

grouped_df = grouped_df[grouped_df['Attrition_Rate'] > 0]

if not grouped_df.empty and grouped_df['Attrition_Rate'].sum() > 0:
    treemap_fig = px.treemap(
        grouped_df,
        path=["LGA_name", "Year"],
        values="Attrition_Rate",
        color="Attrition_Rate",
        color_continuous_scale="RdBu_r",
        title="Treemap of Attrition Rates by LGA and Year"
    )
    treemap_fig.update_layout(
        width=1200,
        height=700,
        plot_bgcolor='white',
    )
    treemap_fig.show()
else:
    print("No valid Attrition Rates to plot.")
```

**TreeMap for A2: Infovis Dataset(NodeLink)**
---



```python
nodes = pd.read_csv("Nodes.csv")
edges = pd.read_csv("Edges.csv")
nodes.head(10)
```

```python
nodes = nodes.rename(columns={"Id": "NodeID", "Label": "NodeName"})
```

```python
edges = edges.merge(nodes, left_on="Source", right_on="NodeID", how="left").rename(columns={"NodeName": "SourceName"}).drop("NodeID", axis=1)

edges = edges.merge(nodes, left_on="Target", right_on="NodeID", how="left").rename(columns={"NodeName": "TargetName"}).drop("NodeID", axis=1)
```

```python
edges.head(10)
```

Heirarchy: Type -> Weight(total nodes connected)


```python
edges['Weight_Sum'] = edges.groupby(['Type', 'SourceName'])['Weight'].transform('sum')

treemap_fig = px.treemap(
    edges,
    path=["Type", "SourceName"],
    values="Weight",
    color="Weight_Sum",
    color_continuous_scale= "viridis",
    title="Tree Map of Types and Weights by Source"
)

treemap_fig.update_layout(
    width=1200,
    height=700,
    plot_bgcolor='white',
)

treemap_fig.show()
```
